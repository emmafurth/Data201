```python
hardcover_alts = df[df["bookformat"].str.contains("Hardcover", na=False) | df["bookformat"].str.contains("Hardback", na=False)]["bookformat"].unique() 
print(hardcover_alts)
goodreads_df = df.replace(to_replace=hardcover_alts, value="Hardcover")

print(goodreads_df["bookformat"].unique())
```
